#  Чистка и добавление служебных колонок товаров с otc.ru


```python
df = pd.read_json('inn-product-all.json')
df = df[df["inn"] != 0]
tmp = df["inn"].apply(lambda x: x in d)
df = df[tmp]
df['okved'] = df['okved'].apply(lambda x: x.replace('ok=', ''))
df.head()
```

```python
df['ogrn'] = df['inn'].apply(lambda x: d[x])
df.columns
```

```python
df = df.rename(columns = {'okved': 'okpd-2'})
clean_columns = ['name', 'okpd-2', 'price',
       'shipping_region', 'ogrn', 'inn', 'shortName']
df = df[clean_columns]
df.head()
```

```python
import re
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('DeepPavlov/rubert-base-cased-sentence')
def _name_clearing(name):
    name = name.replace("\n", "")
    name = re.sub('[^а-яА-Я\s]', '', name)
    name = name.lstrip()
    name = name.rstrip()
    if "  " in name:
        name = name[:name.find("  ")]
    return name.lower()

def _del_enters(x, col_name):
    s = x[col_name]
    if isinstance(s ,str):
        return s.lstrip().rstrip()
    return s

def _convert_price(x):
    if '/' in x:
        x = x[:x.find('/')-2]
    x = ''.join([a for a in x if a in ',.0123456789'])
    x = x.replace(' ', '')
    x = x.replace(',', '.')
    if len(x) == 0: return None

    return float(x)
    
def process_data(df):
    df["name"] = df.apply(lambda x:_del_enters(x, "name"), axis = 1)
    df["shipping_region"] = df.apply(lambda x:_del_enters(x, "shipping_region"), axis = 1)
    df["sys_float_price"] = df["price"].apply(_convert_price)
    df["sys_simple_name"] = df["name"].apply(_name_clearing)
    df["sys_embbedings"] = list(model.encode(df["sys_simple_name"].values))
    return df

df = process_data(df)
df.head()
```

```python
df['description'] = df['name'].copy()
df.head()
```

```python
df.to_csv('otc_okveds.csv', index = False)
```

# Добавление информации из общего реестра 


```python
okved_info = pd.read_csv("okved_info_otc.csv")
okved_info = okved_info.dropna(subset=["ogrn"])
okved_info["ogrn"] = okved_info["ogrn"].apply(int)
okved_info["ogrn"] = okved_info["ogrn"].apply(str)

okved_info.head()
```

```python
s1 = set(df['ogrn'].values)
s2 = set(okved_info['ogrn'].values)
s = list(s1 & s2)
mask = okved_info["ogrn"].apply(lambda x: x in s)
okved_info = okved_info[mask]
mask = df["ogrn"].apply(lambda x: x in s)
df = df[mask]

len(df[mask])/len(df)
```

```python
def _get_reliability(x, okved_info):
    try: 
        reliability = okved_info[okved_info["ogrn"] == x["ogrn"]]["reliability"].values[0]
    except: 
        return "нет информации"
    if isinstance(reliability, str):
        return reliability
    else:
        return "нет информации"

df['reliability'] = df.apply(lambda x: _get_reliability(x, okved_info), axis = 1)
```
