# Мердж поставщиков

```python
import json

berezka_supp = pd.read_json("berezka_supp_table.json")
otc_supp = pd.read_json("otc_supp_table.json")
```

```python
supp = pd.concat([berezka_supp, otc_supp], axis = 0)
supp = supp.drop_duplicates()
```

```python
supp_c = set(supp["ogrn"])
prod_c = set(prod["ogrn"])
ogrns = list(prod_c & supp_c)
```

```python
mask = supp["ogrn"].apply(lambda x: x in ogrns)
supp = supp[mask]
mask = prod["ogrn"].apply(lambda x: x in ogrns)
prod = prod[mask]
```

```python
prod = prod.reset_index()
supp = supp.reset_index()

prod.to_json("products.json")
supp.to_json("suppliers.json")
```

```python
with open('supp_name_dict_otc.json') as f:
    d1 = json.load(f)
with open('supp_name_dict_b.json') as f:
    d2 = json.load(f)
len(d1)
```

```python
d1
```

```python
d1.update(d2)
len(d1)
```

```python
def create_name(x, d):
    if x in d:
        return d[x]
    else:
        return "ООО 'ДЕФЕНС'"
supp['company_name'] = supp["ogrn"].astype(str).apply(lambda x: create_name(x,d1))
```

```python
def str_to_list(x):
    x = str(x)
    x = x.replace("'", '"')
    try:
        return json.loads(x)
    except:
        return None
supp["additionalokveds"] = supp["additionalokveds"].apply(str_to_list)
```

```python
prod = prod.reset_index(drop=True)
supp = supp.reset_index(drop=True)

prod.to_json("products.json")
supp.to_json("suppliers.json")
```
