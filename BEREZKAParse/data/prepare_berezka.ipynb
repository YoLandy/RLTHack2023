```python
import pandas as pd
import numpy as np
import json
```

#  Чистка и добавление служебных колонок товаров с berezka.ru


```python
df = pd.read_csv('berezka_data.csv', index_col = 0)

df.head()
```

```python
df = df.dropna(subset = ["supplierOgrn"])
df = df.reset_index()
```

```python
dict_file = 'supp_name_dict_b.json'
supp_name_dict = {str(int(df["supplierOgrn"][x])): df["supplierName"][x] for x in range(len(df))}
supp_name_dict
```

```python
with open(dict_file, 'w') as file:
    json.dump(supp_name_dict, file)
```

```python
df = df.rename(columns = {'product_name': 'name',
                            'supplierInn': 'inn',
                          'supplierOgrn': 'ogrn',
                         'product_name': 'name', 
                         'supplierInn': 'inn',
                          'okpd2Codes': 'okpd-2',
                          
                          
                        })
df.columns
```

```python
clean_columns = ['name', 'okpd-2', 'price',
        'ogrn', 'inn', 'description', 'supplierName']
df = df[clean_columns]
```

```python
#fix okpd-2
def fix_okpd(x):
    if isinstance(x, str):
        x = x.replace("'",'"')
        x = json.loads(x)
        code = x[0]['code']
        return code
    return x

df['okpd-2'] = df['okpd-2'].apply(fix_okpd)
df.head()
```

```python
import math
tmp = df["ogrn"].apply(lambda x: math.isnan(x))
df = df[~tmp]
df["ogrn"] = df["ogrn"].apply(int)
df.head()
```

```python
def _fill_description(x):
    if isinstance(x['description'], str):
        return x['description']
    else:
        return x['name']
df['description'] = df.apply(_fill_description, axis =1)
```

```python
df = df.drop_duplicates()
df.head()
```

# Добавление информации из общего реестра 

```python
okved_info = pd.read_csv('okved_info_berezka.csv')
okved_info.head()
```

```python
s1 = set(df['ogrn'].values)
s2 = set(okved_info['ogrn'].values)
s = list(s1 & s2)
mask = okved_info["ogrn"].apply(lambda x: x in s)
okved_info = okved_info[mask]
mask = df["ogrn"].apply(lambda x: x in s)
df = df[mask]

len(df[mask])/len(df)
```

```python
def _get_shipping_region(x, okved_info):
    try:
        address = okved_info[okved_info["ogrn"] == x["ogrn"]]["address"].values[0]
    except:
        return "нет информации"
    if isinstance(address, str):
        address = address.split()
        return address[0] + ' ' + address[1]
    else:
        return "нет информации"
    
df['shipping_region'] = df.apply(lambda x: _get_shipping_region(x, okved_info), axis = 1)
```

```python
def _get_reliability(x, okved_info):
    reliability = okved_info[okved_info["ogrn"] == x["ogrn"]]["reliability"].values[0]
    if isinstance(reliability, str):
        return reliability
    else:
        return "нет информации"

df['reliability'] = df.apply(lambda x: _get_reliability(x, okved_info), axis = 1)
```

# Добавление служебных колонок и выделение эмбедингов

```python
import re
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('DeepPavlov/rubert-base-cased-sentence')
def _name_clearing(name):
    name = str(name)
    name = name.replace("\n", "")
    name = re.sub('[^а-яА-Я\s]', '', name)
    name = name.lstrip()
    name = name.rstrip()
    if "  " in name:
        name = name[:name.find("  ")]
    return name.lower()

def _del_enters(x, col_name):
    s = x[col_name]
    if isinstance(s ,str):
        return s.lstrip().rstrip()
    return s

def _convert_price(x):
    if '/' in x:
        x = x[:x.find('/')-2]
    x = ''.join([a for a in x if a in ',.0123456789'])
    x = x.replace(' ', '')
    x = x.replace(',', '.')
    if len(x) == 0: return None

    return float(x)
    
def process_data(df):
    df["name"] = df.apply(lambda x:_del_enters(x, "name"), axis = 1)
    df["shipping_region"] = df.apply(lambda x:_del_enters(x, "shipping_region"), axis = 1)
    
    df["sys_float_price"] = df["price"].apply(float)
    df["sys_simple_name"] = df["name"].apply(_name_clearing)
    df["sys_embbedings"] = list(model.encode(df["sys_simple_name"].values))
    return df

df = process_data(df)
df.head()
```

```python
df.to_json('berezka_products.json')
```
