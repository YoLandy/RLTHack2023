```python
def prepare_items(response):
    products = []
    products_raw = response.get('items', None)
    img_url_start = 'https://offers-api.agregatoreat.ru/api/image/'

    if products_raw != None and len(products_raw) > 0:
        for product in products_raw:
            if product.get('price', None) != None and \
               product.get('price', None) > 10:
                img_id = product.get('imageId', None)
                if img_id == None:
                    img_url = None
                else:
                    img_url = img_url_start + img_id
                    
                products.append({
                    'id': product.get('id', None),
                    'okpd2Codes': product.get('okpd2Codes', None),
                    'product_name': product.get('name', None),
                    'price': product.get('price', None),
                    'img_url': img_url
                })

    return products

def parse_products(book):
    ans = pd.DataFrame()
    for page in range(1, book['num_pages']+1):
        
        response = get_category(book['okpd2'], page)
        products = prepare_items(response)
        
        df = pd.DataFrame(products)
        ans = pd.concat([ans, df], axis=0, ignore_index=True)

    return ans
```
